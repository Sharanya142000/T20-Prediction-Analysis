## Calculating mod_player batting and bowling impacts based on previous season

```python
def calculate_mod_player_batting_impact(name,season):
    if season == 2008:
        return 0.0
    else:
        if len(df_player_stats_new[(df_player_stats_new['fullName'] == name) & (df_player_stats_new['season'] == season-1)]) == 1:
            return df_player_stats_new[(df_player_stats_new['fullName'] == name) & 
                    (df_player_stats_new['season'] == season-1)]['Batting_Player_Impact'].values[0]
        else:
            return np.nan
        
        
def calculate_mod_player_bowling_impact(name,season):
    if season == 2008:
        return 0.0
    else:
        if len(df_player_stats_new[(df_player_stats_new['fullName'] == name) & (df_player_stats_new['season'] == season-1)]) == 1:
            return df_player_stats_new[(df_player_stats_new['fullName'] == name) & 
                    (df_player_stats_new['season'] == season-1)]['Bowling_Impact_Score'].values[0]
        else:
            return np.nan
```

```python
df_player_stats_new['mod_batting_player_impact'] = df_player_stats_new.apply(
    lambda x : calculate_mod_player_batting_impact(x['fullName'],x['season']),axis=1)
```

```python
df_player_stats_new['mod_bowling_player_impact'] = df_player_stats_new.apply(
    lambda x : calculate_mod_player_bowling_impact(x['fullName'],x['season']),axis=1)
```

## Imputing the NULL values with the means

```python
df_player_stats_new["mod_batting_player_impact"] = df_player_stats_new["mod_batting_player_impact"].fillna(
                df_player_stats_new.groupby('fullName')['Batting_Player_Impact'].transform('mean'))

df_player_stats_new["mod_batting_player_impact"] = df_player_stats_new["mod_batting_player_impact"].apply(
                                                                lambda x : round(x,2))
```

```python
df_player_stats_new["mod_bowling_player_impact"] = df_player_stats_new["mod_bowling_player_impact"].fillna(
                df_player_stats_new.groupby('fullName')['Bowling_Impact_Score'].transform('mean'))

df_player_stats_new["mod_bowling_player_impact"] = df_player_stats_new["mod_bowling_player_impact"].apply(
                                                                lambda x : round(x,2))
```

## Calculating relative batting and bowling impacts based on cumulative impacts

```python
def calculate_rel_batting_impact_score(name):
    impact_scores = df_player_stats_new[df_player_stats_new['fullName'] == name].sort_values(by='season')['mod_batting_player_impact'].values
    cum_score = impact_scores[0]
    cum_impact_score_list = [impact_scores[0]]
    for i in range(1,len(impact_scores)):
        if(impact_scores[i] < impact_scores[i-1]):
            cum_score -= (impact_scores[i-1] - impact_scores[i])
        else:
            cum_score+=impact_scores[i]
        cum_impact_score_list.append(round(cum_score,2))
    return cum_impact_score_list
```

```python
total_impact_scores_bat = []
for name in df_player_stats_new['fullName'].unique():
    cum_impact_score_list = calculate_rel_batting_impact_score(name)
    total_impact_scores_bat.extend(cum_impact_score_list) 
    
```

```python
def calculate_rel_bowling_impact_score(name):
    impact_scores = df_player_stats_new[df_player_stats_new['fullName'] == name].sort_values(by='season')['mod_bowling_player_impact'].values
    cum_score = impact_scores[0]
    cum_impact_score_list = [impact_scores[0]]
    for i in range(1,len(impact_scores)):
        if(impact_scores[i] < impact_scores[i-1]):
            cum_score -= (impact_scores[i-1] - impact_scores[i])
        else:
            cum_score+=impact_scores[i]
        cum_impact_score_list.append(round(cum_score,2))
    return cum_impact_score_list
```

```python
total_impact_scores_bowl = []
for name in df_player_stats_new['fullName'].unique():
    cum_impact_score_list = calculate_rel_bowling_impact_score(name)
    total_impact_scores_bowl.extend(cum_impact_score_list) 
```
