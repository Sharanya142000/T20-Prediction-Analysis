```python
import pandas as pd
df=pd.read_csv('IPL 2008-2022 data/all_season_bowling_card.csv')
```

```python
df
```

```python
df['MatchPlayed']=1
df['balls_bowled']=1
```

```python
for i in range(len(df)):
    if df.loc[i]['overs']==0.0 or  df.loc[i]['overs']==1.0 or df.loc[i]['overs']==2.0 or df.loc[i]['overs']==3.0 or df.loc[i]['overs']==4.0:
        df.at[i,'balls_bowled']=df.loc[i]['overs']*6
        continue
    if df.loc[i]['overs']>=0.1 and df.loc[i]['overs']<=0.5:
        df.at[i,'balls_bowled']=df.loc[i]['overs']*10
        continue
    if df.loc[i]['overs']>=1.1 and df.loc[i]['overs']<=1.5:
        df.at[i,'balls_bowled']=6+((df.loc[i]['overs']-1)*10)
        continue
    if df.loc[i]['overs']>=2.1 and df.loc[i]['overs']<=2.5:
        df.at[i,'balls_bowled']=12+((df.loc[i]['overs']-2)*10)
        continue
    if df.loc[i]['overs']>=3.1 and df.loc[i]['overs']<=3.5:
        df.at[i,'balls_bowled']=18+((df.loc[i]['overs']-3)*10)
        continue
```

```python
import numpy as np
df['4W_haul']=np.where(df['wickets']==4,1,0)
df['5W_haul']=np.where(df['wickets']==5,1,0)
```

```python
bowling=df.groupby(['fullName','season'])['balls_bowled','conceded','wickets','MatchPlayed','4W_haul','5W_haul','dots','foursConceded','sixesConceded','wides','noballs','maidens'].sum().reset_index()
bowling
```

```python
bowling['total_overs_bowled']=(bowling['balls_bowled']//6)+((bowling['balls_bowled']%6)/10)
bowling
```

```python
bowling['economy']=np.where(bowling['total_overs_bowled']>0,round(bowling['conceded']/bowling['total_overs_bowled'],2),0)
bowling['strike_rate']=np.where(bowling['wickets']>0,round(bowling['balls_bowled']/bowling['wickets'],2),0)
bowling['average']=np.where(bowling['wickets']>0,round(bowling['conceded']/bowling['wickets'],2),0)
```

```python
bowling
```

```python
bowling['Quality']=np.where(bowling['strike_rate']*bowling['economy']>0,round(100/(bowling['economy']*bowling['strike_rate']),2),0)
bowling['Frequency']=np.where(bowling['MatchPlayed']>0,round(bowling['balls_bowled']/bowling['MatchPlayed'],2),0)
bowling['Bowling_Player_Impact']=round(bowling['Quality']*bowling['Frequency'],2)
bowling['4W_factor']=np.where(bowling['MatchPlayed']>0,round(bowling['4W_haul']/bowling['MatchPlayed'],2),0)
bowling['5W_factor']=np.where(bowling['MatchPlayed']>0,round(bowling['5W_haul']/bowling['MatchPlayed'],2),0)
bowling['Bowling_Impact_Score']=round(bowling['Bowling_Player_Impact']+(bowling['Bowling_Player_Impact']*bowling['5W_factor'])+(bowling['Bowling_Player_Impact']*bowling['4W_factor']/2),2)
bowling
```

```python
bowling.rename(columns={'strike_rate':'Bowling_strike_rate','Quality':'Bowling_Quality','Frequency':'Bowling_Frequency'},inplace=True)
```

```python
bowling.to_csv('Bowling_Features_IPL.csv',index=False)
```
