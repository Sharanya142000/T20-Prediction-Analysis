```python
import pandas as pd
df=pd.read_csv('IPL 2008-2022 data/all_season_batting_card.csv')
```

```python
df
```

```python
import numpy as np

df['Matches']=np.where(df['ballsFaced']>0,1,np.where((df['ballsFaced']==0) & (df['isNotOut']==False),1,0))
```

```python
df.loc[df['ballsFaced']==0,['runs','ballsFaced','isNotOut','Matches']]
```

```python

df['out']=np.where(df['isNotOut']==False,1,0)
```

```python
df['Fifties']=np.where((df['runs']>=50)&(df['runs']<=99),1,0)
df['Hundreds']=np.where(df['runs']>=100,1,0)
```

```python
batting_stats=df.groupby(['fullName','season'])['runs','ballsFaced','out','Matches','fours','sixes','Fifties','Hundreds'].sum().reset_index()
batting_stats
```

```python
batting_stats['batting_avg']=np.where(batting_stats['out']>0,round(batting_stats['runs']/batting_stats['out'],2),0)
```

```python
batting_stats['strike_rate']=np.where(batting_stats['ballsFaced']>0,round(100*batting_stats['runs']/batting_stats['ballsFaced'],2),0)
batting_stats['Quality']=round((batting_stats['batting_avg']+batting_stats['strike_rate'])/160,2)
batting_stats['Frequency']=np.where(batting_stats['Matches']>0,round(batting_stats['ballsFaced']/batting_stats['Matches'],2),0)
batting_stats['Batting_Player_Impact']=round(batting_stats['Quality']*batting_stats['Frequency'],2)
batting_stats['100_Factor']=np.where(batting_stats['Matches']>0,round(batting_stats['Hundreds']/batting_stats['Matches'],2),0)
batting_stats['50_Factor']=np.where(batting_stats['Matches']>0,round(batting_stats['Fifties']/batting_stats['Matches'],2),0)
batting_stats['Batting_Impact_Score']=round(batting_stats['Batting_Player_Impact']+(batting_stats['Batting_Player_Impact']*batting_stats['100_Factor'])+(batting_stats['Batting_Player_Impact']*batting_stats['50_Factor']/2),2)
batting_stats
```

```python
batting_stats.rename(columns={'strike_rate':'Batting_strike_rate','Quality':'Batting_Quality','Frequency':'Batting_Frequency'},inplace=True)
```
