## Mapping player impact scores according to their name and season

```python
def retrieve_player_impact_score(season,x):
    player_impact_score = []
    for player in x:
        if len(df_player_stats_final.query(f'fullName == "{player}" and season == {season}')) == 1:
            player_impact_score.append(round(sum(df_player_stats_final.query(f'fullName == "{player}" and season == {season}').values[0].tolist()[2:]),2))
        else:
            player_impact_score.append(0)
    return player_impact_score
```

```python
df_train['home_player_impact_score'] = df_train.apply(lambda x : retrieve_player_impact_score(x['season'],x['home_playx1']),axis=1)
```

```python
df_train['away_player_impact_score'] = df_train.apply(lambda x : retrieve_player_impact_score(x['season'],x['away_playx1']),axis=1)
```
